## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

I will be choosing scoring and ranking as content items will be assigned a 'performance score' that ranks which signals (position, freshness, content type etc.) are most strongly associated with high engagement rate.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

**Target:** engagement_rate (observed continuous metric)

**Predictions:**
*   Engagement rate: which pages people interact with the most (continuous observed metric)

## 3. Success metric

*One metric you can defend. What number means 'good'?*


*   The primary success metric is Precision@K, where the top K pieces is ranked. For example:

Target: Precision@100 > 0.7 (at least 70 of the top 100 recommendations are correct, which is a good accuracy)

*  A secondary metric is spearman correlation, which tells us how well our ranked signal profile actually correlate with engagement. For example:

Target: r > 0.5 (shows moderate to strong signal pattern, which is reassurance that the pattern is real and not coincidental.)



## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/reezcon/First-ML-Pipeline/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

print("UNIT OF ANALYSIS: ONE ROW = ONE CONTENT PIECE")
print(f"\nTotal content pieces: {len(df):,}")

# Relevant columns for ranking/scoring:
cols_for_ranking = [
    "content_id",
    "engagement_rate",
    "position_tier",
    "freshness_tier",
    "content_type",
    "word_count",
    "cpc",
    "competition_level",
    "ctr"
]

print("\nSample of 5 content pieces (signals + target):")
print(df[cols_for_ranking].head(5))
print("\nTarget distribution (engagement_rate_)")
print(df["engagement_rate"].describe())
print("\nSignal distributions")
print("\nContent type breakdown:")
print(df['content_type'].value_counts())
print("\nPosition tier breakdown:")
print(df['position_tier'].value_counts())
print("\nFreshness tier breakdown:")
print(df['freshness_tier'].value_counts())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Simply using if statements for this model would get too messy because a site's engagement depends on multiple interacting signals (position, freshness, type, word count) and shifts over time. Therefore, it requires the model to learn their true weights and combinations, which is extremely difficult with fixed rules.
